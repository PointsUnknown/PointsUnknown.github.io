```python
parcel = gpd.read_file("Parcels.zip")
```

```python
parcel.head()
```

The selling land data is from [Realtor](https://www.realtor.com/realestateandhomes-search/Fresno_CA/type-farms-ranches) and [Zillow](https://www.zillow.com/homedetails/1784-S-Chateau-Fresno-Ave-Fresno-CA-93706/18665719_zpid/).

```python
filter_list = ["32808006", "32807037", "32907045", "32714063", "33502076S", "32713008","32702127"]
```

```python
sell_farm = parcel[parcel['APN'].isin(filter_list)]
```

```python
parcel[parcel['APN']=='32702127']
```

```python
sell_farm
```

```python
import altair as alt
```

```python
sell_farm.crs
```

```python
sell_farm = sell_farm.to_crs('EPSG:4326')
```

```python
alt.Chart(sell_farm).mark_geoshape().encode(
    tooltip=['APN']
).project(
    type='albersUsa'
).properties(
    title='Selling lands in Fresno',
    width=500,
    height=500
)
```

```python
sell_farm.to_file("sell-farm.geojson", driver='GeoJSON')
```
